# Simple Linear Regression: Predicting Salary from Experience

## 1. Importing Necessary Packages

In this step, we import essential Python libraries for data manipulation, visualization, and machine learning. These packages provide the tools needed to load data, process it, and build regression models.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

## 2. Loading the Dataset and Splitting Features and Target

Here, we load the salary data from a CSV file. We then separate the independent variable (Years of Experience) as X and the dependent variable (Salary) as y. The dataset is split into training and test sets to evaluate the model's performance on unseen data.

```python
data = pd.read_csv('Salary_Data.csv')
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

## 3. Training the Simple Linear Regression Model

In this step, we import the LinearRegression class from scikit-learn and create a regressor object. The model is trained using the training data (X_train and y_train) to learn the relationship between years of experience and salary.

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

## 4. Visualizing the Training Set Results

This step visualizes the results of the linear regression model on the training data. The scatter plot shows the actual data points (red), while the blue line represents the model's predictions. This helps us see how well the model fits the training data.

```python
import matplotlib.pyplot as plt
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

## 5. Visualizing the Test Set Results

Here, we visualize the model's predictions on the test data. The red points are the actual test data, and the blue line is the regression line learned from the training data. This visualization helps us assess how well the model generalizes to new, unseen data.

```python
plt.scatter(X_test, y_test, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
